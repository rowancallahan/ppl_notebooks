# Multiclass Logistic Regression in Pyro
Here is a short notebook to create multiclass logistic regression in pyro.
This will use class plates as I find them a bit easier to look at and read through.
However this could also be done with to_event etc. Remember that for N classes we will do one set of coefficients for each class and get N logits. sets before combining with softmax. The softmax function allows us to take logits calculated for K different classes and combine them such that the probability is normalized correctly. It is smooth and differentiable but it can also be viewed as a

From Wikipedia: 
$$P(y = k | \mathbf{x}) = \frac{e^{\sum_{j=1}^{P} x_j \beta_{jk}}}{\sum_{i=1}^{K} e^{\sum_{j=1}^{P} x_j \beta_{ji}}}$$

The use here of two different summations should give you some kind of clue to the plate structure that I will use below also. I will be summing across all of the covariates and across all the K classes.


For now lets set up some data that we can use to regress on.

```python
import torch
import matplotlib.pyplot as plt
import numpy as np
# Generate 2D separable data
N, K = 150, 3
N_per_class = N // K

x = torch.zeros(N, 2)
y = torch.zeros(N, dtype=torch.long)

# Julia colors: red bottom-left, purple bottom-right, green top
means = torch.tensor([[-2, -2], [2, -2], [0, 2]], dtype=torch.float)

for k in range(K):
   start_idx = k * N_per_class
   end_idx = (k + 1) * N_per_class
   x[start_idx:end_idx] = torch.randn(N_per_class, 2) * 0.5 + means[k]
   y[start_idx:end_idx] = k

# Shuffle
perm = torch.randperm(N)
x = x[perm]
y = y[perm]

# Plot with Julia colors
colors = ['#CB3C33', '#9558B2', '#389826']  # Julia red, purple, green
for k in range(K):
   mask = y == k
   plt.scatter(x[mask, 0], x[mask, 1], c=colors[k], label=f'Class {k}', alpha=0.6)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True, alpha=0.3)
plt.title('3-Class 2D Data')
plt.show()
```

# Creating a three way classifier 
Now that we have our three separable classes we want to calculate a multinomial logistic regression to separate them
 The following I will use for super high dimensional data, but this example holds in just two dimensions
 however since we will later scale this up I will have each dimension also take a plate.

```python
import torch
import torch.nn.functional as F
from torch import nn

import pyro
import pyro.distributions as dist
from pyro.infer import Predictive, SVI, Trace_ELBO
from pyro.infer.autoguide import AutoNormal
from pyro.optim import Adam
import pyro.poutine as poutine
from pyro.nn import PyroModule
from pyro.nn import PyroSample

from pyro.ops import einsum

from tqdm import tqdm
assert pyro.__version__.startswith('1.9.1')
pyro.set_rng_seed(0)
class bayesian_multiclass(PyroModule):
    def forward(self, x, classnum, y=None):
        P = x.size(1)
        K = classnum
  
        class_plate = pyro.plate("class_plate", K, dim=-2)
        with class_plate:
            with pyro.plate("coeff_plate", P, dim=-1):
                _shared = pyro.sample("_shared", dist.Normal(0., 100.))
                _shared = pyro.sample("_shared", dist.InverseGamma(1.,1.))
                c = pyro.sample("coeff", dist.Normal(_shared, _shared))


        score = torch.einsum("...c, kc->...k", x, c)

        #N.B. We are only clamping the score here because the separation is artificial
        #this will cause the predicted likelihoods to have so high of a ratio
        #that they will overflow when exponentiated in the softmax function
        score = torch.clamp(score, min=-10, max=10) 
        probs = F.softmax(score, dim=-2) 
        total_prob = pyro.deterministic("total_prob", probs)

        with pyro.plate("data", x.size(0)): 
            obs = pyro.sample("obs", dist.Categorical(probs=probs), obs=y)

```

# Training the model

Now we instantiate the model, and create a guide for it. I will be using the AutoNormal guide and the Trace_ELBO loss. 
Pyro is pretty good when it comes to making guide especially for basic problems so its a good idea to start there knowing the guide will also be fast. We will run for `1500` steps with a learning rate of `0.005`

```python
optim = Adam({"lr": 0.005})
model = bayesian_multiclass()
guide = AutoNormal(model)
svi = SVI(model, guide, optim, loss=Trace_ELBO())
losses = []
train_num = 1500
def train(num_iterations):
    for j in tqdm(range(num_iterations)):
        loss = svi.step(x,3,y)
        losses.append(loss)
pyro.clear_param_store()
train(train_num)
```

# Examining Training

This looks pretty good now lets quickly plot our training to make sure we saturated and havent under trained the model.


```python
plt.scatter(np.arange(0,train_num,1), losses, alpha=0.08)
plt.plot(np.arange(0,train_num,1), losses, alpha=1.0, lw=0.4)
```

Seeing as we have a nice smooth L, now we can move on and see how our classifications play out on the dataset.

```python
predictive = Predictive(model, guide=guide, num_samples=300)
predicted_values = predictive(x, 3,None)

#Take the predicted values, turn them into a float so we can average
#get the average predicted value, round to whole number then turn into int
labels = predicted_values["obs"].float().mean(dim=0).round().int().numpy()
```

# Take away

Although this might seem like a bit of overkill with use of plate notation and
einstein summation to calculate a logistic regression classifier.
There are some advantages to writing things in this manner. For one, the
approach that we are using here scales very easily to very big P (tons of variables) and
large K. This means that taking this same model we can do a bayesian regression with 100 classes 1000 features
and potentially millions of points. The plate notation and einstein summation also make it a bit easier in my
opinion to add extra features to the model and deal with them, while keeping the plates and dimensions more clear in your head.
